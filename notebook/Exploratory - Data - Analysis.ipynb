## Business Understanding

Before starting the analysis, it is important to understand the purpose of each table.

* **Purchases Table:** Contains all purchase transactions, including the vendor, product, purchase date, quantity purchased, and total purchase amount.
* **Purchase Prices Table:** Stores the selling price and purchase price of each product supplied by every vendor. This table helps calculate product profit margins.
* **Vendor Invoice Table:** Contains invoice details for each purchase order, including freight charges. It is useful for analyzing vendor expenses and payment information.
* **Sales Table:** Stores actual sales transactions, including products sold, sales quantity, sales price, and total sales revenue.

Since the required information is stored in different tables, we need to combine them into a single summary table. This summary table will help us analyze:

* Vendor purchase transactions
* Product sales performance
* Freight cost for each vendor
* Product purchase and selling prices
* Vendor profitability
* Product pricing optimization

Creating a summary table makes the analysis easier and helps generate meaningful business insights for decision-making.

## Vendor Sales Summary

#### Objective

The business data is stored in multiple tables, where each table contains different information related to vendors, products, purchases, sales, and invoices.

To perform vendor performance analysis, we need to combine these tables into a single summary dataset.

This query creates a consolidated **Vendor Sales Summary** by integrating procurement, sales, pricing, and freight information.

---

### Why is this summary table required?

The required business information is spread across multiple tables:

* **Purchases Table** → Purchase transactions
* **Purchase Prices Table** → Product purchase price and selling price
* **Sales Table** → Sales quantity and sales revenue
* **Vendor Invoice Table** → Freight cost

Instead of analyzing each table separately, we create one analytical dataset that contains all the important business metrics.

This makes further analysis easier and improves dashboard performance.

---

#### Step 1: Freight Summary

The first Common Table Expression (CTE) calculates the total freight cost for each vendor.

##### Purpose

* Calculate total freight cost
* Analyze logistics expenses
* Measure vendor transportation cost

**Output**

* Vendor Number
* Total Freight Cost

---

#### Step 2: Purchase Summary

The second CTE combines the **Purchases** table with the **Purchase Prices** table.

#### Purpose

This step calculates:

* Purchase Price
* Selling Price
* Product Volume
* Total Purchase Quantity
* Total Purchase Cost

This provides a complete procurement summary for every vendor and product.

---

#### Step 3: Sales Summary

The third CTE summarizes the sales transactions.

#### Purpose

It calculates:

* Total Sales Quantity
* Total Sales Revenue
* Total Selling Price
* Total Excise Tax

This helps measure the sales performance of each product.

---

#### Step 4: Create Final Vendor Sales Summary

Finally, all three summary tables are joined together using **LEFT JOIN**.

The final dataset contains:

* Vendor Information
* Product Information
* Purchase Details
* Sales Details
* Freight Cost

This becomes the main dataset for business analysis.

---



#### Business Insights

* A single analytical dataset is easier to analyze than multiple tables.
* Combining procurement, sales, pricing, and freight data provides a complete business view.
* This summary table improves reporting efficiency.
* It serves as the foundation for KPI calculation and Power BI dashboard development.

```python
import sqlite3
import pandas as pd
import logging

logging.basicConfig(
    filename="logs/Vendor_Sales_Summary.log",
    level=logging.DEBUG,
    format="%(asctime)s - %(levelname)s - %(message)s",
    filemode="a"
)   
def ingest_db(df, table_name, conn):
    df.to_sql(
        name=table_name,
        con=conn,
        if_exists="replace",
        index=False
    )

    logging.info(f"{table_name} table successfully saved into the database.")

    

def create_vendor_summary(conn):
    Vendor_Sales_Summary = pd.read_sql_query("""WITH FreightSummary AS(
select 
VendorNumber , SUM(Freight) as FreightCost
from vendor_invoice
group BY VendorNumber
),

PurchaseSummary AS (
SELECT
p.VendorNumber,
p.VendorName,
p.Brand,
p.Description,
p.PurchasePrice,
pp.Volume,
pp.Price as ActualPrice,
sum(p.Quantity) as TotalPurchasesQuantity,
sum(p.Dollars) as TotalPurchasesDollars
from purchases p
join purchase_prices pp
on p.Brand = pp.Brand
where p.purchaseprice > 0
group by p.VendorNumber,p.VendorName,p.Brand ,p.Description,p.PurchasePrice,pp.Price,pp.Volume
),

SalesSummary AS (
select 
VendorNo,
Brand,
sum(SalesQuantity) as TotalSalesQuantity,
sum(SalesDollars) as TotalSalesDollars,
sum(SalesPrice) as TotalSalesPrice,
sum(ExciseTax) as TotalExciseTax
from sales
group by VendorNO, Brand 
)

Select 
ps.VendorNumber,
ps.VendorName,
ps.Brand,
ps.Description,
ps.PurchasePrice,
ps.ActualPrice,
ps.Volume,
ps.TotalPurchasesQuantity,
ps.TotalPurchasesDollars,
ss.TotalSalesQuantity,
ss.TotalSalesDollars,
ss.TotalSalesPrice,
ss.TotalExciseTax,
fs.FreightCost

from PurchaseSummary ps
left join SalesSummary ss
on ps.VendorNumber = ss.VendorNo
and ps.Brand = ss.Brand
left join FreightSummary fs
on ps.VendorNumber = fs.VendorNumber
order by ps.TotalPurchasesDollars desc""",conn)
    return Vendor_Sales_Summary
    
  


def clean_data(df):
    """
    Clean the vendor sales summary dataset
    and create business KPIs.
    """

    # Data Type Conversion
    df['Volume'] = df['Volume'].astype(float)

    # Fill Missing Values
    df.fillna(0, inplace=True)

    # Remove Extra Spaces
    df['VendorName'] = df['VendorName'].str.strip()
    df['Description'] = df['Description'].str.strip()

    # Feature Engineering
    df['GrossProfit'] = (
        df['TotalSalesDollars'] -
        df['TotalPurchasesDollars']
    )

    df['ProfitMargin'] = (
        df['GrossProfit'] /
        df['TotalSalesDollars']
    ) * 100

    df['StockTurnover'] = (
        df['TotalSalesQuantity'] /
        df['TotalPurchasesQuantity']
    )

    df['SalesToPurchasesRatio'] = (
        df['TotalSalesDollars'] /
        df['TotalPurchasesDollars']
    )

    return df



if __name__ == "__main__":

    # Create Database Connection
    conn = sqlite3.connect("inventory.db")

    logging.info("Creating Vendor Sales Summary...")

    summary_df = create_vendor_summary(conn)

    logging.info("Cleaning Vendor Sales Summary...")

    clean_df = clean_data(summary_df)

    logging.info("Saving Data into Database...")

    ingest_db(
        clean_df,
        "vendor_sales_summary",
        conn
    )

    logging.info("Vendor Sales Summary saved successfully.")

    conn.close()

    logging.info("Database connection closed.")
```
